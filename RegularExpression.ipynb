```python
# 20141111-2.py
# 正規表達式 Regular Expression
import requests
from bs4 import BeautifulSoup
import re 
```

```python
m = re.match(r"(\w+)@(\w+)", "david@numerinfo.com") 
print m.group(1), m.group(2)
print m.groups()
```

```python
m = re.match(r"(\w+)@([a-z.]+)", "david@numerinfo.com") 
print m.groups() 
```

```python
m = re.match(r"(\d+)\.(\d+)", "1999.5") 
print m.groups() 
```

```python
m = re.match(r"(?P<first_name>\w+) (?P<last_name>\w+)", "David Chiu") 
print m.group('first_name'), m.group('last_name')
```

```python
str1 = 'scp file.txt root@10.0.0.1:./' 
m=re.search(r'^scp ([\w\.]+) (\w+)@([\w\.]+):(.+)',str1) 
if m: 
    print m.group(1), m.group(2), m.group(3), m.group(4)
    print m.groups()
```

```python
num = '0912345678'
m = re.match(r'(\d{10})', num)
print m.group(0)
```

```python
phone = ['0912345678', '0912-345678', '0912-345-678']
for p in phone:
    #print p
    m = re.match(r'(\d{4})-?(\d{3})-?(\d{3})', p)
    print m.groups()
    print m.group(1), m.group(2), m.group(3)
```
